```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("C:\\Users\\deepa\\data-export (1).csv")
```

```python
df.head()
```

```python
df.columns= df.iloc[0]
df= df.drop(index = 0).reset_index(drop = True)
df.columns = ["channel group","DateHour", "Users","Sessions","Engaged sessions",	"Average engagement time per session",	"Engaged sessions per user",	"Events per session",	"Engagement rate",	"Event count"]
```

```python
df.head()
```

```python
df.info()
```

# cleaning data and data validation for the above datas set

```python
df.head()
```

```python
df["DateHour"]= pd.to_datetime(df["DateHour"], format= "%Y%m%d%H", errors='coerce')
                               
```

```python
df.head()
```

```python
numerical_cols= df.columns.drop(["channel group", "DateHour"])
df[numerical_cols]=df[numerical_cols].apply(pd.to_numeric, errors= 'coerce')
df["Hour"]= df["DateHour"].dt.hour
```

```python
df.head()
```

```python
df.info()
```

# session and user over time

```python
sns.set(style= "whitegrid")
```

```python
plt.figure(figsize=(10,6))
df.groupby("DateHour")[["Sessions", "Users"]].sum().plot(ax=plt.gca())
plt.title("Sesions and users over time")
plt.xlabel("DateHour")
plt.ylabel("Count")
plt.show()
```

# Total users by channel

```python
plt.figure(figsize=(10,5))
sns.barplot(data= df, x="channel group", y="Users", estimator= np.sum, palette="viridis")
plt.title("Total users by Channel")
plt.xticks(rotation=45)
plt.show()
```

# Average engagement time by channel and behaviour ,content effectiveness

```python
plt.figure(figsize=(10,5))
sns.barplot(data= df, x="channel group", y="Average engagement time per session", estimator= np.sum, palette="magma")
plt.title("Average engagement time by channel")
plt.xticks(rotation=45)
plt.show()
```

```python
df.head()
```

# Engagement rate distribution by channel

```python
plt.figure(figsize=(10,5))
sns.boxplot(data= df, x="channel group", y="Engagement rate", palette="coolwarm")
plt.title("Engagement rate distribution by channel")
plt.xticks(rotation=45)
plt.show()
```

# traffic by hour and channel


```python
heatmap_data = df.groupby(["Hour","channel group"])["Sessions"].sum().unstack().fillna(0)
plt.figure(figsize=(10,5))
sns.heatmap(heatmap_data,cmap="YlGnBu", linewidth=.5,annot= True,fmt='.0f')
plt.title("traffic by hour and channel")
plt.xlabel("Channel Group")
plt.ylabel("Hour of days")
plt.show()
```

```python
df_plot = df.groupby("DateHour") [["Engagement rate", "Sessions"]].mean().reset_index() 
plt.figure(figsize=(10,5)) 
plt.plot(df_plot ["DateHour"], df_plot ["Engagement rate"], label="Engagement rate", color="green")
plt.plot(df_plot ["DateHour"], df_plot ["Sessions"], label="Sessions", color="blue")
plt.title(" Engagement Rate vs Sessions Over Time")
plt.xlabel("DateHour")
plt.legend()
plt.grid(True)
plt.show()
```
